```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 50)
y1 = x**2
y2 = 2*x + 1
```

```python
# 图，一个figure表示一张图，从一个figure到下一个figure为止
plt.figure(figsize=(8, 5))
# 颜色、线宽度、线的风格（虚实）
plt.plot(x, y1, color='black', linewidth = 5, linestyle='--')
plt.plot(x, y2) 
# 取值范围
plt.xlim(-1, 1)     
plt.ylim(-2,3)
# 轴描述 label
plt.xlabel("scores")
plt.ylabel("comment")
# 坐标轴的数字 ticks（x坐标轴本来是0 0.25 0.5 ---> 0 0.5 1）
new_ticks = np.linspace(-1,1,5)
plt.xticks(new_ticks)
# alpha数学形式打印   r"$ \ $"
plt.yticks(
[-2, -1.8, 0, 1.22, 3],
    [r"$really\ bad$",r"bad",r"$normal\ \alpha$","good","wonderful"]
)
# 改变坐标轴位置
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')    # 让上、右坐标轴消失
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
plt.show()
```
